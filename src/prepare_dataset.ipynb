```python
def truncate(prompts):
    tokenizer = tiktoken.get_encoding("cl100k_base")

    truncate_prompts = []

    for prompt in prompts:
        tokens = tokenizer.encode(prompt)
        truncate_tokens = tokens[0 : np.random.randint(1, len(tokens) - 1)]
        truncate_prompts.append(tokenizer.decode(truncate_tokens))

    return truncate_prompts

def prepare_train_data(prompts, completions):
    train_data = []

    for i in range(len(prompts)):
        train_data.append({"prompt" : prompts[i], "completion" : completions[i]})
    
    fields = ['prompt', 'completion'] 
        
    # name of csv file 
    filename = "data.csv"
        
    # writing to csv file 
    with open(filename, 'w', encoding="UTF-8") as csvfile: 
        # creating a csv dict writer object 
        writer = csv.DictWriter(csvfile, fieldnames = fields) 
            
        # writing headers (field names) 
        writer.writeheader() 
            
        # writing data rows 
        writer.writerows(train_data) 
```
